```python
!pip install transformers

from transformers import BertTokenizer
import torch
from transformers import BertForSequenceClassification

model = BertForSequenceClassification.from_pretrained("bert-base-uncased",
                                                      num_labels=7,
                                                      output_attentions=False,
                                                      output_hidden_states=False)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

#model.to(device)
```

```python
model.load_state_dict(
    torch.load('/content/drive/MyDrive/Colab Notebooks/Berta/SA.model',
              map_location=torch.device('cpu'))
)
```

```python
label_to_word = {
  1: 'Joy',
  2: 'Fear',
  3: 'Anger',
  4: 'Sadness',
  5: 'Disgust',
  6: 'Ashamed',
  7: 'Guilt'
}

def sa (intxt):
  encoded = tokenizer(intxt, return_tensors='pt')
  logits = model(**encoded).logits
  predicted_class_id = logits.argmax().item()
  #print(logits)
  print(label_to_word[predicted_class_id+1])
```

```python
sa('issues like affordable housing, climate action, Skytrain to UBC, and more are at stake.')
```
